# Error bars


```python
import matplotlib.pyplot as plt

fig,ax=plt.subplots()
movie_type=['Comedy','action','Romance','Drama','Scifi']
Response=[4,5,6,1,4]
ax.bar(movie_type,Response, yerr=True)
ax.set_ylabel('Response',color='r',fontsize=15)
ax.set_xticklabels(movie_type,Rotation=75)
plt.show()


#Line indicates that where the majority of the responses is available.
```

```python
import matplotlib.pyplot as plt
import statistics as St

fig,ax=plt.subplots()
movie_type=['Comedy','action','Romance','Drama','Scifi']
Response=[4,5,6,1,4]
mean= St.mean(Response)
std= St.stdev(Response)
ax.bar(movie_type,Response, yerr=mean)
ax.set_ylabel('Response',color='r',fontsize=15)
ax.set_xticklabels(movie_type,Rotation=75)
plt.show()

fig.savefig("Movie.png",dpi=300)
```

```python
import matplotlib.pyplot as plt

fig,ax=plt.subplots()
movie_type=['Comedy','action','Romance','Drama','Scifi']
Men_Response=[4,5,6,1,4]
women_response=[8,2,7,2,1]

ax.bar(movie_type,Men_Response,color='r',width=0.25)
ax.bar(movie_type,women_response,color='g',width=0.25)

ax.set_ylabel('Response',color='r',fontsize=16)
ax.set_xticklabels(movie_type,Rotation=45)
ax.legend(['MEN','WOMEN'])
```

```python
#stacked bar chart

import matplotlib.pyplot as plt

fig,ax=plt.subplots()
movie_type=['Comedy','action','Romance','Drama','Scifi']
Men_Response=[4,5,6,1,4]
women_response=[8,2,7,2,1]

ax.bar(movie_type,Men_Response,color='r',width=0.25)
ax.bar(movie_type,women_response,bottom=Men_Response,color='g',width=0.25)

ax.set_ylabel('Response',color='r',fontsize=16)
ax.set_xticklabels(movie_type,Rotation=45)
ax.legend(['MEN','WOMEN'])
```

```python
#BOX PLOTS

import numpy as np

np.random.seed(10)

col1=np.random.normal(100,10,200)
col2=np.random.normal(90,20,200)
col3=np.random.normal(80,30,200)
col4=np.random.normal(70,40,200)
data=[col1,col2,col3,col4]

fig=plt.figure()

ax=fig.add_axes([0,0,1,1])

ax.boxplot(data)

plt.show()
```

```python
#Violin plot


import numpy as np

np.random.seed(10)

col1=np.random.normal(100,10,200)
col2=np.random.normal(90,20,200)
col3=np.random.normal(80,30,200)
col4=np.random.normal(70,40,200)
data=[col1,col2,col3,col4]

fig=plt.figure()

ax=fig.add_axes([0,0,1,1])

ax.violinplot(data)

plt.show()
```
