```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
#scatter plot
x=[23,6,79,35,47,80,42]
y=[12,18,22,28,32,38,55]

sb.scatterplot(x,y)

plt.show()
```

```python
#count plot

gender=['Male','Female','Female','Female','Male','Male','Female']

sb.countplot(x=gender)

plt.show()
```

```python
df=sb.load_dataset('tips')
df
```

```python
#scatter plot with hue

sb.scatterplot(x='total_bill',y='tip',hue='smoker',hue_order=['Yes','No'],data=df)
```

```python
sb.scatterplot(x='total_bill',y='tip',hue='smoker',data=df)
```

```python
hue_col={'Yes':'Blue','No':'Red'}

sb.scatterplot(x='total_bill',y='tip',data=df,hue='smoker',hue_order=['Yes','No'],palette=hue_col)
```

```python
#count plot
hue_col={'Yes':'Blue','No':'Red'}
sb.countplot(x='smoker',hue='sex',data=df)
```

```python
#Relation plot

sb.relplot(x='total_bill',y='tip',data=df,kind='scatter')
```

```python
#subplots

sb.relplot(x='total_bill',y='tip',data=df,kind='scatter',col="smoker")
sb.relplot(x='total_bill',y='tip',data=df,kind='scatter',col="sex")
```

```python
#subplots

sb.relplot(x='total_bill',y='tip',data=df,kind='scatter',col="smoker",row='sex')

```

```python
sb.relplot(x='total_bill',y='tip',data=df,kind='scatter',col='smoker',row='day')
```

```python
#wrapping and ordering colomns

sb.relplot(x='total_bill',y='tip',data=df,kind='scatter',col='day')
sb.relplot(x='total_bill',y='tip',data=df,kind='scatter',col='day',col_wrap=2)
```

```python
#SIZE AND HUE

sb.relplot(x='total_bill',y='tip',data=df,kind='scatter',size='size',hue='size',alpha=0.9)
```

```python
#line plots

sb.relplot(x='total_bill',y='tip',data=df,kind='line')
```

```python
sb.relplot(x='total_bill',y='tip',data=df,kind='line',style='day',hue='day',markers=True,dashes=False)
```

```python
sb.relplot(y='total_bill',x='day',data=df,kind='scatter')
```

```python
sb.relplot(y='total_bill',x='day',data=df,kind='line')
```

```python
sb.relplot(y='total_bill',x='day',data=df,kind='line',ci='sd') #shaded region is called confidence interval ci
```

```python
sb.relplot(y='total_bill',x='day',data=df,kind='line',ci=None)
```

```python
#count plot and bar plots
#subplots

sb.catplot(x='smoker',data=df,kind='count')

```
